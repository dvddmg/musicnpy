```python
from musicnpy import *
from IPython.display import Image
```

```python
a = Scale([0, 1, 2, 3, 4, 5, 6], 60)
b = a.sort('<').shift(-9)
c = b.interleave(a)
c -= 23
a = Staff(a.values).out
b = Staff(b.values).out
c = Staff(c.values).out

Score((a, b, c), filename='./images/scale', format='png', size=(150, 100)).make_file
Image(filename='./images/scale.png')
```

```python
a = Scale.new(PMod.minNat, 60)
b = Scale.new(PMod.maj, 48)
c = Scale.new(PMod.maj, 72)

b *= 2

c = (c*0.5) + a.offset

d = c.interpolation(a, 7, 4)
e = _Set(d[-1]).interpolation(b, 7, 2)



# questo perch pyly non supporta al momento i quarti di tono
# values da ottimizzare rispetto a pyly
d = Staff([x.round(decimals=0).values for x in d]).out
e = Staff([x.round(decimals=0).values for x in e]).out

# print(d)
Score((d, e), filename='./prova', format='png').make_file
Image(filename='./prova.png')
```
